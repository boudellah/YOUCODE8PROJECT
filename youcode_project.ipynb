```python
new_row = {'Pays': 'Greece' , ' Age ': 25, 'Genre': 'Male', 'Dépenses des clients': 100}
new_row2 = {'Pays': 'Italy', ' Age ': 30, 'Genre': 'Female', 'Dépenses des clients':200}

# Ajouter la nouvelle ligne au DataFrame Pays
df.loc[len(df)] = new_row 
df.loc[len(df)] = new_row2
```

```python

# Calculer la moyenne de la colonne Âge
mean_age = df[' Age '].mean()

# Remplacer l'âge à l'index 2 par la moyenne
df.loc[45, ' Age '] = int(mean_age)
```

```python
# Check the data types in the column
print(df['Dépenses des clients'].dtype)
```

```python
# Check for non-numeric values in the column
non_numeric_values = df[~df['Dépenses des clients'].apply(lambda x: str(x).replace('.', '', 1).isdigit())]
print("Non-numeric values in 'Dépenses des clients':")
print(non_numeric_values['Dépenses des clients'])
```

```python
# Convert the 'Dépenses des clients' column to numeric, forcing errors to NaN
df['Dépenses des clients'] = pd.to_numeric(df['Dépenses des clients'], errors='coerce')

# Now, you can calculate the mean
mean_spending = df['Dépenses des clients'].mean()

print(f"Mean spending: {mean_spending}")
```

```python
# Calculer la médiane et la moyenne pour les colonnes spécifiées
median_age = df[' Age '].median()
mean_age = df[' Age '].mean()

median_spending = df['Dépenses des clients'].median()
mean_spending = df['Dépenses des clients'].mean()

# Afficher les résultats
print(f"Médiane de l'âge: {median_age}")
print(f"Moyenne de l'âge: {mean_age}")

print(f"Médiane des dépenses des clients: {median_spending}")
print(f"Moyenne des dépenses des clients: {mean_spending}")
```

# Partie 2

```python
import matplotlib.pyplot as plt
```

```python
# Regrouper les données par pays et calculer la somme des dépenses par pays
expenses_by_country = df.groupby('Pays')['Dépenses des clients'].sum()

# Créer le graphique à barres
plt.figure(figsize=(10, 6))
expenses_by_country.plot(kind='bar', color='skyblue')
plt.title('Dépenses des clients par pays')
plt.xlabel('Pays')
plt.ylabel('Dépenses des clients (€)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Nettoyer les lignes avec moins de 10 € de dépenses
df_cleaned = df[df['Dépenses des clients'] >= 10]
```

```python
# Nettoyer les doublons
df_cleaned = df_cleaned.drop_duplicates()
```

```python
# Écrire les données nettoyées dans un nouveau fichier CSV
df_cleaned = df_cleaned[['Pays', ' Age ', 'Genre', 'Dépenses des clients']]
df_cleaned.to_csv('fichier_nettoye.csv', index=False)
```
